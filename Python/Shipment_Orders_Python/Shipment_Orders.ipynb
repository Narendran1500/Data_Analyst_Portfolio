```python
# considering the values of Not Available , unknown as na values
ship_data = pd.read_excel('C:/Users/NARENDRAN/Downloads/Python/Shipment_orders_data/orders_data.xlsx',na_values=['Not Available', 'unknown'])
ship_data['Ship Mode'].unique()
```

```python
ship_data.head()
```

```python
ship_data.rename(columns={"Order Id":"Order_Id",
                         "Order Date":"Order_Date",
                         "Ship Mode":"Ship_Mode",
                         "Postal Code":"Postal_Code",
                         "Sub Category":"Sub_Category",
                         "Product Id":"Product_Id",
                         "cost price":"Cost_Price",
                         "List Price":"List_Price",
                         "Discount Percent":"Discount_Percent"},inplace=True)
ship_data.info()
```

```python
# adding a new column selling price
ship_data['Selling_Price']= ship_data['List_Price'] - ship_data['List_Price']*(ship_data['Discount_Percent']/100)
ship_data['Selling_Price']
```

```python
ship_data.head(3)
```

```python
# adding profit column
ship_data['Profit']= ship_data['Selling_Price'] - ship_data['Cost_Price']
ship_data['Profit']
```

```python
ship_data.head()
```

```python
# renaming the columns
ship_data.rename(columns={"Profit" : "Unit_Profit",
                         "Selling_Price" : "Unit_Selling_Price"},inplace= "True")
ship_data.head()
```

```python
ship_data.dtypes
```

# Dropping few columns

```python
ship_data.drop(columns = ['Cost_Price','List_Price','Discount_Percent'],inplace=True)
```

```python
ship_data.head()
```

```python
ship_data['Total_Profit'] = ship_data['Quantity']*ship_data['Unit_Profit']
ship_data.head(2)
```

### Find month over month growth comparison for 2022 and 2023

```python
ship_data.head(2)
```

```python
ship_pivot = ship_data.groupby(['Month','Year'])[['Total_Sales']].sum().reset_index().pivot_table(index='Month', columns='Year', values='Total_Sales', aggfunc='sum').reset_index()
ship_pivot
```

```python
ship_pivot['YoY_growth']= ship_pivot[2022] - ship_pivot[2023]
ship_pivot
```

### For each category which month had highest sales

```python
monthly_sales = ship_data.groupby(['Month','Category'])[['Total_Sales']].sum().reset_index()
highest_sales_month = monthly_sales.loc[monthly_sales.groupby('Category')['Total_Sales'].idxmax()]
highest_sales_month
```
